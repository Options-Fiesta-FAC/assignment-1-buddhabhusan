```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#strike price
k = input("enter your strike price ")
#Premium
p = input("enter your premium ")
#spot price
s_now = input("enter your current mkt price ")

#Converting string to float
s_now = float(s_now)
k = float(k)
p = float(p)
```

Defining function

```python
def buy_call(k,p,s):
  payoff = np.maximum(s-k,0) - p
  return payoff

def sell_call(k,p,s):
  payoff = p - np.maximum(s-k,0)
  return payoff

def buy_put(k,p,s):
  payoff = np.maximum(k-s,0) -p
  return payoff

def sell_put(k,p,s):
  payoff = p - np.maximum(k-s,0)
  return payoff
```

```python
print(f"If we buy a call {buy_call(k,p,s_now)} is the payoff")
print(f"If we sell a call {sell_call(k,p,s_now)} is the payoff")
print(f"If we buy a put {buy_put(k,p,s_now)} is the payoff")
print(f"If we sell a put {sell_put(k,p,s_now)} is the payoff")
```

```python
#generating spot price for plotting graph 
s = np.linspace(0,2*s_now , 100)

#payoff based on the buying or selling of the option
buy_c_payoff = buy_call(k,p,s)
sell_c_payoff = sell_call(k,p,s)
buy_p_payoff = buy_put(k,p,s)
sell_p_payoff = sell_put(k,p,s)
```

In Longing a  call opiton we get profit only after ours spot price is greater than strike price + premium (break even).<BR>
Profit Potential = Unlimited <br>
Loss Potential = Limited(premium)<br>
<br>
From the Graph we can sat that if the spot price goes above the beakeven, i will start getting profit. Otherwise i will be in loss.


```python

plt.figure(figsize=(8,6))
plt.plot(s, buy_c_payoff, color='green', label='Buy Call')
plt.title("Long Call")
plt.xlabel("spot price")
plt.ylabel("payoff")
plt.axhline(0, color='black', linewidth=0.5)  # horizontal line at 0
plt.axvline(k, color='red', linestyle='--', label="Strike Price (K)")
plt.axvline(k+p, color='black', linestyle='--', label="Break_even")
plt.legend()
plt.show()
```

In Shorting a  call opiton we get profit only after ours spot price is less than strike price (break even).<BR>
Profit Potential = Limited(premium)<br>
Loss Potential =  Unlimited <br>
<br>
From the Graph we can sat that if the spot price goes abov the beakeven, i will start getting profit. Otherwise i will be in loss.


```python
plt.figure(figsize=(8,6))
plt.plot(s, sell_c_payoff, color='green', label='Buy Call')
plt.title("Short Call")
plt.xlabel("spot price")
plt.ylabel("payoff")
plt.axhline(0, color='black', linewidth=0.5)  # horizontal line at 0
plt.axvline(k, color='red', linestyle='--', label="Strike Price (K)")
plt.axvline(k+p, color='black', linestyle='--', label="Break_even")
plt.legend()
plt.show()
```

In longing a put opiton we get profit only after ours spot price is less than strike price (break even).<BR>
Profit Potential = Limited(premium), it is the premium the buyer have to pay for the right of call option<br>
Loss Potential =  Unlimited , as the stock price can get too high.<br>
<br>
From the Graph we can sat that if the spot price goes below the beakeven, i will start getting profit. Otherwise i will be in loss.


```python

plt.figure(figsize=(8,6))
plt.plot(s, buy_p_payoff, color='green', label='Buy Call')
plt.title("Long PUT")
plt.xlabel("spot price")
plt.ylabel("payoff")
plt.axhline(0, color='black', linewidth=0.5)  # horizontal line at 0
plt.axvline(k, color='red', linestyle='--', label="Strike Price (K)")
plt.axvline(k-p, color='black', linestyle='--', label="Break_even")
plt.legend()
plt.show()
```

In shorting a put opiton we get profit only after ours spot price is more than strike price (break even).<BR>
Profit Potential = Unlimited , as the stock price can get too high the buyer will pay the seller as he had the rights when the market was going up<br>
Loss Potential = Limited(premium), it is the premium the seller gets for giving the right of this option to buyer.<br>
<br>
From the Graph we can sat that if the spot price goes abov the beakeven, i will start getting profit. Otherwise i will be in loss.


```python
plt.figure(figsize=(8,6))
plt.plot(s, sell_p_payoff, color='green', label='Buy Call')
plt.title("Short Put")
plt.xlabel("spot price")
plt.ylabel("payoff")
plt.axhline(0, color='black', linewidth=0.5)  # horizontal line at 0
plt.axvline(k, color='red', linestyle='--', label="Strike Price (K)")
plt.axvline(k-p, color='black', linestyle='--', label="Break_even")
plt.legend()
plt.show()
```
